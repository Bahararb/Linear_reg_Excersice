```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df1 = df.rename(columns = {"per capita income (US$)" : "Price"})
df1
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price per $') 
plt.scatter(df1.year, df1.Price, color='green')
```

```python
new_df = df1.drop('Price' ,axis='columns')
new_df
```

```python
price = df1.Price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[2009]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
